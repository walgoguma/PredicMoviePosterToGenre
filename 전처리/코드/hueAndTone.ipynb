```python
import cv2
import numpy as np

def ColorSystem(RGB):
    oneBitImg=np.uint8([[RGB]])
    hsv=cv2.cvtColor(oneBitImg,cv2.COLOR_RGB2HSV)
    
    # h: 색상, s:채도, v:명도
    h=hsv[0][0][0]
    s=hsv[0][0][1]
    v=hsv[0][0][2]
    
    result=[]
    
    '''    
    R 빨강 (Red) YR 주황 (Yellow Red) Y 노랑 (Yellow) GY 연두 (GreenYellow)
    G 녹색 (Green) BG 청록 (Blue Green) B 파랑 (Blue)
    PB 남색 (Purple Blue) P 보라 (Purple) RP 자주 (Red Purple)
    '''
    #180/10
    if (18*0<=h)and(18*1>=h):
        result.append('R')
    elif (18*1<h)and(18*2>=h):
        result.append('YR')
    elif (18*2<h)and(18*3>=h):
        result.append('Y')
    elif (18*3<h)and(18*4>=h):
        result.append('GY')   
    elif (18*4<h)and(18*5>=h):
        result.append('G')     
    elif (18*5<h)and(18*6>=h):
        result.append('BG')      
    elif (18*6<h)and(18*7>=h):
        result.append('B')       
    elif (18*7<h)and(18*8>=h):
        result.append('PB')       
    elif (18*8<h)and(18*9>=h):
        result.append('P')
    elif (18*9<h)and(18*10>=h):
        result.append('RP')
    elif h==0:
        result.append('RPrp')    
    else:
        print("error h:")
        print(h)
    

    '''
    세로열 _ V 산뜻한 (Vivid) S 기본색 (Strong) B 밝은 (Bright) P 해맑은 (Pale)
    VP 아주 연한 (Very Pale) Lgr 밝은 회 (Light Grayish) L 옅은 (Light)
    Gr 회 (Grayish) DI 짙은 (Dull) DP 어두운 (Deep) DK 어두운 회 (Dark)
    '''
    
    if (256/22*0<=s)and(256/22*1>=s):
        result.append('V')
    elif (256/22*1<s)and(256/22*3>=s):
        result.append('S')
    elif (256/22*3<s)and(256/22*4>=s):
        result.append('B')
    elif (256/22*4<s)and(256/22*5>=s):
        result.append('P')   
    elif (256/22*5<s)and(256/22*7>=s):
        result.append('Vp')     
    elif (256/22*7<s)and(256/22*9>=s):
        result.append('Lgr')      
    elif (256/22*9<s)and(256/22*12>=s):
        result.append('L')       
    elif (256/22*12<s)and(256/22*14>=s):
        result.append('Gr')       
    elif (256/22*14<s)and(256/22*17>=s):
        result.append('Dl')
    elif (256/22*17<s)and(256/22*19>=s):
        result.append('Dp')
    elif (256/22*19<s)and(256/22*22>=s):
        result.append('DK')
    '''
    N 무채색 (Netural) _ 흰색(N9) / 밝은 회색 / 회색 / 어두운 회색 / 검정(N1)
    '''    
    if (256/9*0<=v)and(256/9*1>=v):
        result.append('N1')
    elif (256/9*1<v)and(256/9*2>=v):
        result.append('N2')
    elif (256/9*2<v)and(256/9*3>=v):
        result.append('N3')
    elif (256/9*3<v)and(256/9*4>=v):
        result.append('N4')   
    elif (256/9*4<v)and(256/9*5>=v):
        result.append('N5')     
    elif (256/9*5<v)and(256/9*6>=v):
        result.append('N6')     
    elif (256/9*6<v)and(256/9*7>=v):
        result.append('N7')      
    elif (256/9*7<v)and(256/9*8>=v):
        result.append('N8')       
    elif (256/9*8<v)and(256/9*9>=v):
        result.append('N9')
    return result
```
