```python
temp=[]
genre=[[] for i in range(0,len(loadData.imdbId))]

for x in range(0,len(loadData.imdbId)):
    temp.append(loadData.Genre[x].split('|'))
    #print(temp[x])
    #print(len(temp[x]))
    for i in range(0,len(temp[x])):
        #print(temp[x][i])
        if (temp[x][i]=='Romance'or temp[x][i]== 'Comedy' or temp[x][i]== 'Thriller' or temp[x][i]== 'Horror' or temp[x][i]== 'Mystery' or temp[x][i]== 'Action'):
            #스릴러, 호러, 미스터리물은 서스펜스로 통합
            if (temp[x][i]== 'Thriller' or temp[x][i]== 'Horror' or temp[x][i]== 'Mystery'):
                genre[x].append('Suspense')
            else:
                genre[x].append(temp[x][i])

            #만약 복합 장르가 아니면
            if(len(genre[x])!=1):
                genre[x].clear()
            
    #print(genre[x])
```
