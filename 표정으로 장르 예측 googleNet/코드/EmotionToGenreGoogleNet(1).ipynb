```python
# from sklearn.datasets import fetch_mldata
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import io
```

```python
loadData = pd.read_csv("C:/Users/eun51/Desktop/emotionData.csv"
                   , encoding = "ISO-8859-1"
                   , engine='python'
                   , keep_default_na=False)
loadData.head()
```

```python
genre=loadData.singleGenre
print(genre)
```

```python
emotion=loadData.singleEmotion
print(emotion)
```

```python
emotionNum=[]
for x in range(0,len(genre)):
    if(emotion[x]=='angry'):
        emotionNum.append([1, 0, 0, 0, 0, 0, 0])
    elif(emotion[x]=='disgust'):
        emotionNum.append([0, 1, 0, 0, 0, 0, 0])
    elif(emotion[x]=='fear'):
        emotionNum.append([0, 0, 1, 0, 0, 0, 0])
    elif(emotion[x]=='happy'):
        emotionNum.append([0, 0, 0, 1, 0, 0, 0])
    elif(emotion[x]=='sad'):
        emotionNum.append([0, 0, 0, 0, 1, 0, 0])
    elif(emotion[x]=='surprise'):
        emotionNum.append([0, 0, 0, 0, 0, 1, 0])
    elif(emotion[x]=='neutral'):
        emotionNum.append([0, 0, 0, 0, 0, 0, 1])
```

```python
print(emotionNum)
print(len(emotionNum))
```

```python
for x in range(0,len(genre)):
    genre[x]=genre[x].replace("[","").replace("]","").replace("''","")
```

```python
print(genre)
```

```python
genreNum=[]
#Romance:0,Comedy:1,Suspense:2,Action:3
for x in range(0,len(genre)):
    if(genre[x]=="'Romance'"):
        genreNum.append([1,0,0,0])
    elif(genre[x]=="'Comedy'"):
        genreNum.append([0,1,0,0])
    elif(genre[x]=="'Suspense'"):
        genreNum.append([0,0,1,0])
    elif(genre[x]=="'Action'"):
        genreNum.append([0,0,0,1])
```

```python
print(genreNum)
print(len(genreNum))
```

```python
# rescale the data, use the traditional train/test split
X_train, X_test = emotionNum[:int(len(emotion)*0.8)], emotionNum[int(len(emotion)*0.8):]
y_train, y_test = genreNum[:int(len(genre)*0.8)], genreNum[int(len(genre)*0.8):]
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
model= MLPClassifier(hidden_layer_sizes=(100, 50),max_iter=100,solver='sgd',learning_rate_init=0.5)
```

```python
model.fit(X_train, y_train)
```

```python
plt.plot(model.loss_curve_)
plt.show()
```

```python
print("Training set score: %f" % model.score(X_train, y_train))
print("Test set score: %f" % model.score(X_test, y_test))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras import metrics
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt

X_train, X_test = emotionNum[:int(len(emotion)*0.8)], emotionNum[int(len(emotion)*0.8):]
y_train, y_test = genreNum[:int(len(genre)*0.8)], genreNum[int(len(genre)*0.8):]

model = Sequential([
    Dense(10, input_shape=(7,)),
    Activation('relu'),
    Dense(4),
    Activation('softmax'),
])

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=[metrics.categorical_accuracy])

history = model.fit(X_train, y_train, epochs=10, batch_size=500)
model.save('EmotionToGenreGoogleNet.h5')
```

```python
history.history
```

```python
#오차 및 정확도 확인하기
fig, axs = plt.subplots(2, 1, figsize=(15,15))
axs[0].plot(history.history['loss'])
axs[0].title.set_text('Training Loss vs Validation Loss')
axs[0].set_xlabel('Epochs')
axs[0].set_ylabel('Loss')
axs[0].legend(['Train','Val'])
axs[1].plot(history.history['categorical_accuracy'])
axs[1].title.set_text('Training Accuracy vs Validation Accuracy')
axs[1].set_xlabel('Epochs')
axs[1].set_ylabel('Accuracy')
axs[1].legend(['Train', 'Val'])
```

```python
score = model.evaluate(X_train, y_train, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1]*100)
```
